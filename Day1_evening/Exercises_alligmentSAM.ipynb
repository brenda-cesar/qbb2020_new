# Number of alligments

```python
file = open("debug.sam", "r")
```

```python
file
```

```python
all_lines = [] 
for rows in file: 
    columns = rows.split('\t') 
    all_lines.append(columns) 
```

```python
for actual_line in all_lines: 
    print(actual_line[9]) 
```

```python
for index, item in enumerate(all_lines): 
    print('Row : {0} Alignment : {1}'.format(index + 1, item[9])) 
```
