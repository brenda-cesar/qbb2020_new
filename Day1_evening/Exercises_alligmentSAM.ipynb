# Exercise 1.1 Count number of alignments

```python
#Open file
file = open("/Users/cmdb/qbb2020-answers/day1-evening/debug.sam", "r")
```

```python
#Check file is there
file
```

```python
all_lines = [] #Create list that has nothing in it
```

```python
for rows in file: 
    columns = rows.split('\t') #split columns by tabs
    all_lines.append(columns) #add columns to a list
```

```python
for actual_line in all_lines: #sort through each row in the list "all_lines"
    print(actual_line[9]) #print only the sequence that matched which is in column 9
```

```python
for index, item in enumerate(all_lines): #enumerate will add an index to all lines; for index, item will add the index as an index and the item as the line  
    print('Row : {0} Alignment : {1}'.format(index + 1, item[9])) #index + 1 because counter starts at 0
```

```python
len(all_lines) #number of times there was an alligment
```

# Exercise 1.2 Count number of alignments that match perfectly to the genome

```python
#Open file
file = open("/Users/cmdb/qbb2020-answers/day1-evening/debug.sam", "r")
```

```python
matches = [] #Create list that has nothing in it
perfect = 'NM:i:0'
perfect
```

```python
for row in file: 
    if perfect in row: #sort the ones that have perfect matches
        sequence = row.split('\t') #split columns by tabs
        matches.append(sequence) #add sequences that match to a list named  matches
```

```python
for line in matches: #sort through each row in the list "matches"
    print(line[9]) #print only the sequence that matched which is in column 9
```

```python
len(matches) #number of alligments that were perfect
```

# Exercise 1.4 Calculate average MAPQ score across all reads

```python
#Open file
file = open("/Users/cmdb/qbb2020-answers/day1-evening/debug.sam", "r")
```

```python
all_lines = [] #Create list that has nothing in it
```

```python
for rows in file: 
    columns = rows.split('\t') #split columns by tabs
    all_lines.append(columns) #add columns to a list
```

```python
MAPQ = [] #Create list that has nothing in it
```

```python
for actual_line in all_lines: #sort through each row in the list "all_lines"
    MAPQ.append(float(actual_line[4])) #append to the list only the MAPQ values which are in row 5 and make them integers
```

```python
MAPQ
```

```python
average = sum(MAPQ)/len(MAPQ) #averageMAPQ
average
```

# Exercise 1.5 Count number of reads that start their alignment on chromosome 2L between base 10000 and 20000 (inclusive)

```python
#Open file
file = open("/Users/cmdb/qbb2020-answers/day1-evening/debug.sam", "r")
```

```python
matches_ch = [] #Create list for matches that are in chromosome 2L
matches_dis = [] #Create list for matches that are between 10000 and 20000
chromosome = '2L'
```

```python
for row in file: 
    if chromosome in row: #sort the ones that have perfect matches
        sequence = row.split('\t') #split columns by tabs
        matches_ch.append(sequence[3]) #add only the start alligment number to a list named  matches_ch
```

```python
matches_ch
```

```python
for line in matches_ch:
    if int(line) > 10000 and int(line) < 20000:
        matches_dis.append(line) #append only those that are between 10000 and 20000
```

```python
matches_dis
```
