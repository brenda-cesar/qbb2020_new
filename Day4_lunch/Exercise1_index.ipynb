```python
raw_data = open("BDGP6.Ensembl.81.gtf","r")
```

```python
raw_data
```

```python
#Eliminate anything that is not in chromosome 3R or is not protein coding
TargChrom = '3R'
biotype = 'protein_coding'
sorted_file = []
```

```python
for line in raw_data:
    if TargChrom not in line:
        continue
    if biotype not in line:
        continue
    if TargChrom and biotype in line:
        new_line = line.split("\t") #becomes a list
        part = new_line[8]
        new_line1 = part.split(";")
        sorted_file.append([new_line[0],new_line[2],new_line[3],new_line[4],new_line1[0]])
```

```python
#print(sorted_file)
```

```python
feature = 'gene'
sorted_list = []
```

```python
for line in sorted_file:
    if line[1] == feature:
        sorted_list.append(line)
        
```
